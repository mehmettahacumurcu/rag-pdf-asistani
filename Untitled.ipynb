# Sıfırdan RAG pipeline'ının oluşturulması

## RAG Nedir ?

RAG -> Retrieval Augmented Generation.

RAG'in amacı bir bilgi kaynağını LLM'e vererek LLM'in bu bilgiler kapsamında çıktı vermesidir.

LLM -> Large Language Model

* Retrieval - Arama: Verilen sorguyla ilişkili bilginin bulunması. 
* Augmented - Eklenmiş (Arttırılmış): Girdimizi (promtumuzu) alakalı bilgiyle eklenmiş hale getirmek.
* Generation - Oluşturma: İlk iki adımdaki işlemlerle oluşturulan girdiyi bir LLM'e vererek üretken bir çıktı almak.

RAG sisteminin tanıtıldığı makale: https://arxiv.org/abs/2005.11401 (Facebook AI)

## Neden RAG kullanıyoruz ?

RAG kullanmanın temel amacı LLM'in çıktı oluşturma sürecinin iyileştirilmesi.

1. Halisünasyonları Engellemek - LLM'ler olasılıksal yapılarından ötürü iyi (gözüken*) çıktılar vermekte oldukça iyiler fakat çıktının gerçek bilgiyle doğrudan bağlantısı yok. RAG yapısı kullanılan modelin bilgiyle doğrudan bağlantısı olmasını sağlar.

2. Özel Veriyle Çalışmak - LLM'ler genel olarak internette mevcut olan veriyle eğitiliyorlar. Bu sebeple genel olarak girilen bir girdinin nasıl bir çıktıya karşılık geldiğini biliyorlar. Spesifik olarak özel bir bilgiye dair tahmin yapma yetenekleri genel bilgiye oranla düşük. RAG özel bilgiye dair çıktının iyileştirilmesini sağlıyor.

## Kullanım alanları

* Müşteri destek servisleri - Şirket verileriyle oluşturulan bir RAG yapısıyla şirketlere özel cevaplar oluşturulması. 
* Birbiriyle alakalı mail analizi - Bir şirket olduğunuzu ve içeride haberleşmenin sağlandığı mail ağı olduğunu var sayalım. Bu ağda spesifik olarak aranan bir bilgiyi sadece LLM kullanarak bulamayacağımız için RAG sistemi kurarak modelden bir asistan yaratma.
* Herhangi bir konudaki notlara dair bilgisi olan bir asistan yaratma.

```python
import os
import requests

pdf_path = "human-nutrition-text.pdf"

if not os.path.exists(pdf_path):
    print(f"[INFO] Dosya bulunamadı. İndiriliyor...")

    url = "https://github.com/mrdbourke/simple-local-rag/blob/main/human-nutrition-text.pdf"

    filename = pdf_path

    response = requests.get(url)

    if response.status_code == 200:
        with open(filename,"wb") as file:
            file.write(response.content)
        print(f"[INFO] Dosya indirildi, {filename} adıyla kayıt edildi.")
    else:
        print(f"[INFO] Dosya indirme başarısız. Durum kodu {response.status_code}")
else:
    print(f"Dosya {pdf_path} mevcut.")
```

```python
import fitz
from tqdm.auto import tqdm

def text_formatter(text: str) -> str:
    """ Performs minor formatting on text."""

    cleaned_text = text.replace("\n" , " ").strip()

    return cleaned_text

def open_and_read_pdf(pdf_path: str) -> list[dict]:
    doc = fitz.open(pdf_path)
    pages_and_text = []

    for page_number , page in tqdm(enumerate(doc)):
        text = page.get_text()
        text = text_formatter(text = text)

        pages_and_text.append({"page_number" : page_number - 41, # BURA ÖNEMLİ PDFLE ALAKLI -41 DEĞİŞTİRMEK GEREKEBİLİR
                              "page_char_count": len(text),
                              "page_word_count": len(text.split(" ")),
                              "page_sentence_count_raw" : len(text.split(". ")),
                              "page_token_count": len(text) / 4, # 1 token ~= 4 chars
                              "text":text}) 
    return pages_and_text

pages_and_texts = open_and_read_pdf(pdf_path = pdf_path)
pages_and_texts[2:]
```

```python
num_sentence_chunk_size = 10


def split_list(input_list: list[str],
              slice_size: int = num_sentence_chunk_size) -> list[list[str]]:
    return [input_list[i:i+slice_size] for i in range (0 , len(input_list) , slice_size)]

test_list = list(range(25))
split_list(test_list)
```

### Her yığını tek bir obje haline getirmek

Her yığını bir vektör olarak ifade etmek istiyoruz. Bu yaklaşım her yığın başındaki anlam-ayrıntı yoğunluğunu korumak için.



```python
import re

pages_and_chunks = []

for item in tqdm(pages_and_texts):
    for sentence_chunk in item["sentence_chunks"]:
        chunk_dict = {}
        chunk_dict["page_number"] = item["page_number"]

        # Cümleleri paragraf yapısına getirmek
        joined_sentence_chunk = "".join(sentence_chunk).replace("  "," ").strip()
        joined_sentence_chunk = re.sub(r'\.([A-Z])' , r'. \1' , joined_sentence_chunk) # BURAYI YAZMAMIZIN SEBEBİ: Textleri chunklara ayırırken önceden kullandığımız birleştirme sebebiyle cümle analizindeki nokta ve sonrasında gelen boşluk kuralı doğru çalışmıyor.
                                                                                       # REGEX'İN YAPTIĞI ŞEY: NOKTADAN SONRA GELEN BİR BÜYÜK HARF VAR İSE NOKTAYLA SONRADAN GELEN BÜYÜK HARF ARASINA BOŞLUK EKLİYOR  ".A" => ". A"  
        chunk_dict["sentence_chunk"] = joined_sentence_chunk


        chunk_dict["chunk_char_count"] = len(joined_sentence_chunk)
        chunk_dict["chunk_word_count"] = len([word for word in joined_sentence_chunk.split(" ")])
        chunk_dict["chunk_token_count"] = len(joined_sentence_chunk) / 4

        pages_and_chunks.append(chunk_dict)

len(pages_and_chunks)
```

Semantik arama pipeline'ının kurulması

Adımlar:

1. Sorgu stringinin oluşturulması.
2. Sorgu stringinin vektörize edilmesi.
3. Dot product* veya kosinüs benzerliği yoluyla sorgu ile text vektörleri arasında benzerliğe bak.
4. Adım 3'teki sonuçları azalır biçimde sırala.

Dot productun kullanılabilmesi için: Vektor boyutları aynı olmalı , vektör veri tipleri aynı olmalı.

```python
query = "macronutrient functions"
print(f"Query: {query}")

# Sorgunun vektörize edilmesi
# Sorgunun vektörize edildiği modelin, bilgilerin vektörize edildiği model ile aynı olması gerekiyor
query_embedding = embedding_model.encode(query , convert_to_tensor = True)

# Benzerlik skoruna bak dot product ile (eğer modelin çıktısı normalize değilse kosinüs benzerliği ile)
from time import perf_counter as timer

start_time = timer()
dot_scores = util.dot_score(a = query_embedding , b = embeddings)[0]
end_time = timer()

print(f"{len(embeddings)} vektör üzerinde skorlandırılmanın yapılması {end_time - start_time} saniye sürdü.")


# En iyi k skorun alınması
top_results_dot_product = torch.topk(dot_scores , k = 5)
top_results_dot_product
```

## Benzerlik Ölçütleri: DOT PRODUCT VE KOSİNÜS BENZERLİĞİ


Bu görsel, bilgisayarın o karmaşık 768 boyutlu vektörleri nasıl tek bir "Benzerlik Skoruna" dönüştürdüğünün en sade ispatıdır. Sadece aynı anlama (aynı yöne) gelen özellikler skoru yükseltir, alakasız (dik) özellikler sonucu etkilemez.

```python
embeddings[0] # 768 Dimension
```

```python
import torch 

def dot_product(vector1 , vector2):
    return torch.dot(vector1, vector2)

def cosine_similarity(vector1 , vector2): # DOT PRODUCT DİVİDED BY EUCLIDEAN NORM
    dot_product = torch.dot(vector1,vector2)

    # Euclidean / L2 Norm - Büyüklüğü karşılaştırmada denklemden çıkarıp sadece dimensionların benzerliğine bakmak için.
    norm_vector1 = torch.sqrt(torch.sum(vector1**2))
    norm_vector2 = torch.sqrt(torch.sum(vector2**2))

    return dot_product / (norm_vector1 * norm_vector2)


vector1 = torch.tensor([1,2,3] , dtype = torch.float32)
vector2 = torch.tensor([1,2,3] , dtype = torch.float32)
vector3 = torch.tensor([4,5,6] , dtype = torch.float32)
vector4 = torch.tensor([-1,-2,-3] , dtype = torch.float32)

print(f"Vektör1 ile vektör2'nin dot product'ı: {dot_product(vector1 , vector2)}")
print(f"Vektör1 ile vektör3'nin dot product'ı: {dot_product(vector1 , vector3)}")
print(f"Vektör1 ile vektör4'nin dot product'ı: {dot_product(vector1 , vector4)}")
print("-"*30)
print(f"Vektör1 ve vektör2 arasındaki kosinüs benzerliği: {cosine_similarity(vector1,vector2)}")
print(f"Vektör1 ve vektör3 arasındaki kosinüs benzerliği: {cosine_similarity(vector1,vector3)}")
print(f"Vektör1 ve vektör4 arasındaki kosinüs benzerliği: {cosine_similarity(vector1,vector4
                                                                            )}")
    
```

## SEMANTİK ARAMA PIPELINE'ININ OLUŞTURULMASI

```python
def retrieve_relevant_resources(query: str,
                               embeddings : torch.tensor,
                               model: SentenceTransformer = embedding_model,
                               n_resources_to_return: int = 5,
                               print_time: bool = True):
    """
    Girilen bir sorguyu vektörize hale getirir, en benzer n adet skoru ve indisi döndürür.
    """

    # Sorgunun vektörize edilmesi
    query_embedding = model.encode(query , convert_to_tensor = True)

    # Dot product skorunun alınması - KOSİNÜS BENZERLİĞİ YERİNE DOT PRODUCT KULLANMAMIZIN SEBEBİ EMBEDDİNG MODELİN HALİ HAZIRDA NORMALİZE BİR ÇIKTI ÜRETMESİ
    start_time = timer()
    dot_scores = util.dot_score(query_embedding , embeddings)[0]
    end_time = timer()

    # if print_time:
        # print(f"[INFO] {len(embeddings)} vektör için değerlendirmenin yapılması {end_time-start_time:.4f} saniye sürdü.")

    scores , indices = torch.topk(input = dot_scores,
                                 k = n_resources_to_return)

    return scores, indices

def print_top_results_and_scores(query: str,
                                embeddings: torch.tensor,
                                pages_and_chunks: list[dict] = pages_and_chunks,
                                n_resources_to_return: int = 5):
    """
    Sorguyla alakalı paragrafları bulur, skorlarıyla birlikte ekrana yazdırır.
    """

    scores, indices = retrieve_relevant_resources(query = query ,
                                                 embeddings = embeddings,
                                                 n_resources_to_return = n_resources_to_return)


    for score , idx in zip(scores , indices):
        print(f"Skor: {score:.4f}")
        print("Text:")
        print_wrapped(pages_and_chunks[idx]["sentence_chunk"])
        print(f"Sayfa numarası: {pages_and_chunks[idx]['page_number']}")
        print('\n')
```

### LLM Kurulumu

1. Quantization config (opsiyonel) - modelin yüklenirken kullanacağı float precision* (hassasiyet - keskinlik -> 8bit , 4bit) değerini belirten konfigürasyon bilgisi.
2. Model ID - modelin kimliği
3. Tokenizer - girdiyi sayılara çeviren kısım (embedding model değil) anlam bilgisi yok.




```python
from transformers import AutoTokenizer , AutoModelForCausalLM , BitsAndBytesConfig
from transformers.utils import is_flash_attn_2_available

# Quantization config -> kullanmak için !pip install bitsandbytes accelerate
quantization_config = BitsAndBytesConfig(load_in_4bit = True,
                                        bnb_4bit_compute_dtype = torch.float16)

# flash attention 2 = attention mekanizması için hızlandırma | Ekran kartınızın hesaplama kapasitesi >= 8.0 olması gerekiyor.
if (is_flash_attn_2_available()) and (torch.cuda.get_device_capability(0)[0]) >= 8:
    attn_implementation = "flash_attention_2"
else:
    attn_implementation = "sdpa" # scaled dot product attention

tokenizer = AutoTokenizer.from_pretrained(pretrained_model_name_or_path = model_id)

llm_model = AutoModelForCausalLM.from_pretrained(pretrained_model_name_or_path = model_id,
                                                 torch_dtype = torch.float16,
                                                 quantization_config = quantization_config if use_quantization_config else None,
                                                 low_cpu_mem_usage = False,
                                                 attn_implementation = attn_implementation)

if not use_quantization_config:
    llm_model.to("cuda")
```

```python
def get_model_num_params(model: torch.nn.Module):
    return sum([param.numel() for param in model.parameters()])

get_model_num_params(llm_model)
```

```python
def get_model_mem_size(model: torch.nn.Module):
    mem_params = sum([param.nelement() * param.element_size() for param in model.parameters()])
    mem_buffers = sum([buf.nelement() * buf.element_size() for buf in model.buffers()])

    model_mem_bytes = mem_params + mem_buffers
    model_mem_mb = model_mem_bytes / (1024**2)

    model_mem_gb = model_mem_bytes / (1024**3)

    return {"model_mem_bytes" : model_mem_bytes,
           "model_mem_mb" : round(model_mem_mb,2),
           "model_mem_gb" : round(model_mem_gb,2)}

get_model_mem_size(llm_model)
```

```python
def prompt_formatter(query: str,
                    context_items: list[dict]) -> str:
    context = "- " + "\n- ".join(item["sentence_chunk"] for item in context_items)



    # İLK CÜMLENİN ALTINDA BUNLAR VARDI
    # If the answer is explicitly in the context, extract it.
    # If the answer is not explicitly in the context, generate with related information from context.
    base_prompt  = """You are a helpful assistant. Use the provided context to answer the question.


Context:
{context}

Question: {query}

Answer:"""

    base_prompt = base_prompt.format(context = context , query = query)

    dialogue_template = [
        {"role":"user",
        "content": base_prompt}
    ]

    prompt = tokenizer.apply_chat_template(conversation = dialogue_template,
                                          tokenize = False,
                                          add_generation_prompt = True)

    
    return prompt

query = random.choice(query_list)
print(f"Sorgu: \n {query}")

scores , indices = retrieve_relevant_resources(query = query,
                                              embeddings = embeddings)

context_items = [pages_and_chunks[i] for i in indices]

prompt = prompt_formatter(query = query , context_items = context_items)

print(prompt)
```

```python
def ask(query: str,
       temperature: float = 0.4,
       max_new_tokens: int = 256,
       format_answer_text = True,
       return_answer_only = True):
    """
    Bir sorgu alır, ilgili bilgileri bulur ve bu bilgiler üzerinden bir cevap üretir
    """

    # RETRIEVAL - BİLGİ ÇEKME
    # skorlar ve indisleri - alakalı bilgiler
    scores , indices = retrieve_relevant_resources(query = query , embeddings = embeddings)

    context_items = [pages_and_chunks[i] for i in indices]

    for i , item in enumerate(context_items):
        item["score"] = scores[i].cpu()

    # AUGMENTATİON
    prompt = prompt_formatter(query = query,
                             context_items = context_items)

    # GENERATION
    input_ids = tokenizer(prompt , return_tensors="pt").to("cuda")

    outputs = llm_model.generate(**input_ids,
                                temperature = temperature,
                                do_sample = True,
                                max_new_tokens = max_new_tokens)

    output_text = tokenizer.decode(outputs[0])

    if format_answer_text:
        output_text = output_text.replace(prompt, "").replace("<bos>" , "").replace("<eos>" , "")

    if return_answer_only:
        return output_text

    return output_text, context_items

    
```
